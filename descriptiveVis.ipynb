```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('finalData.csv')

df = pd.DataFrame(df)
```

```python
df1 = df[(df['season_x'] > 2017)]
```

```python
df2 = df[(df['season_x'] > 2017)]

drivers_to_include = [
    "Joey_Logano", "Denny_Hamlin", "Kyle_Busch", "Martin_Truex_Jr",
    "Brad_Keselowski", "Kyle_Larson", "Ryan_Blaney", "William_Byron"]

df2 = df2[df2['driver_id'].isin(drivers_to_include)]
```

Distribution of Wins By Track Type for Entire Dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

wins_by_track = df[df['finish_position'] == 1].groupby('race_track_surface').size().reset_index(name='wins')

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='race_track_surface', y='wins', data=wins_by_track)
plt.xlabel('Track Type')
plt.ylabel('Number of Wins')
plt.savefig('wins_by_track.png')
plt.show()
```

Distribution of Wins by Track Type for Dataset used for training

```python
wins_by_track2 = df2[df2['finish_position'] == 1].groupby('race_track_surface').size().reset_index(name='wins')

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='race_track_surface', y='wins', data=wins_by_track2)
plt.xlabel('Track Type')
plt.ylabel('Number of Wins')
plt.savefig('wins_by_trackType.png')
plt.show()
```

Distribition of Wins by Track Length for entire Dataset

```python
wins_by_track_length = df[df['finish_position'] == 1].groupby('race_lap_len').size().reset_index(name='wins')

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='race_lap_len', y='wins', data=wins_by_track_length)
plt.xlabel('Track Type')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.savefig('wins_by_trackLength.png') 
plt.show()
```

Distribution of Wins by Track Length for NN training set

```python
wins_by_track_length2 = df2[df2['finish_position'] == 1].groupby('race_lap_len').size().reset_index(name='wins')

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='race_lap_len', y='wins', data=wins_by_track_length2)
plt.title('Distribution of Wins by Track Length')
plt.xlabel('Track Type')
plt.ylabel('Number of Wins')
plt.show()
```

Distribution of Wins by each Driver in the NN Training Set

```python
wins_by_driver = df2[df2['finish_position'] == 1].groupby('driver_id').size().reset_index(name='wins')

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='driver_id', y='wins', data=wins_by_driver)
plt.xlabel('Driver Name')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.savefig('wins_by_driverSample.png', bbox_inches='tight')
plt.show()
```

Distribution of Wins by Track for each driver in the NN Training Set

```python
wins_by_track = df2[df2['finish_position'] == 1].groupby('track_name').size().reset_index(name='wins')

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='track_name', y='wins', data=wins_by_track)
plt.xlabel('Track')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.savefig('wins_by_trackSample.png', bbox_inches='tight')
plt.show()
```
